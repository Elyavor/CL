```python
import json
import string
import re
# !pip install pymorphy2
import numpy
import pandas as pd
import pymorphy2
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
```

```python
analyzer = pymorphy2.MorphAnalyzer()
```

```python
file = open("reviews1.json", mode='r', encoding='utf-8') 
read_file = file.read()
# print(read_file)
```

## Предобработка текста

```python
read_file = read_file.replace('"',"'")
read_file = read_file.replace('«'," ")
read_file = read_file.replace('»'," ")
read_file = read_file.replace("'title': '",'"title": "')
read_file = read_file.replace("'description': '",'"description": "')
read_file = read_file.replace("'\n",'"\n')
read_file = read_file.replace("',\n",'",\n')
read_file = read_file.replace(".",'. ')
read_file = read_file.replace("ё",'е')
read_file = read_file.replace('Ё','Е')
read_file = read_file.replace("!",' ')
read_file = read_file.replace("…",' ')
read_file = read_file.replace("©",' ')
read_file = read_file.replace("—",' ')
read_file = read_file.replace("“",' ')
read_file = read_file.replace("”",' ')
```

```python
corpus = json.loads(read_file)
```

```python
def preprocess(text):
    text = re.sub("[\W ]+", ' ', text)
    tokens = word_tokenize(text)
    tokens = [token.lower() for token in tokens]
    return tokens
```

```python
def lemmatize(w):
    return analyzer.parse(w)[0].normal_form
```

```python
def filter_noun(doc):
    doc = [w for w in doc if analyzer.parse(w)[0].tag.POS == "NOUN"]
    return doc
```

```python
def remove_stops(doc):
    stops = stopwords.words("russian")
    doc = [w for w in doc if w not in stops]
    return doc
```

```python
docs = []
for i in corpus:
    print("Запускаем предоработку...")
    doc = preprocess(i['description'])
    print("Фильтруем существительные...")
    doc = filter_noun(doc)
    print("Удаляем стоп-слова...")
    doc = remove_stops(doc)
    print("Лемматизируем текст...")
    doc = [lemmatize(w) for w in doc]
    docs.append(doc)
print(docs)
```

```python
filter_noun(docs[0])
```

## Топ-50 слов

```python
from collections import Counter
```

```python
word_dict = Counter()

for sent in docs:
    counter = Counter(sent)
    word_dict.update(counter)

word_dict = dict(word_dict.most_common(50))
len(word_dict)
```

```python
edges_weights = [] # (слово1, слово2, n), n - сколько предложений содержат оба слова
```

Для слов находим индексы документов, в которых они находится

```python
def det_doc_ids(docs):
    dictionary = {}
    index = 0
    for doc in docs:
        for token in doc:
            if token in dictionary:
                dictionary[token].add(index)
            else:
                dictionary[token] = {index}
                index += 1
    return dictionary

dictionary = det_doc_ids(docs)
# print(dictionary)
```

```python
keys = list(word_dict.keys())
for i, word1 in enumerate(word_dict): 
    for j in range(i+1, len(word_dict)):
        word2 = keys[j]
        cnt = dictionary[word1] & dictionary[word2] # получаем множества id док-ов для слов, берем пересечение
        edges_weights.append((word1, word2, len(cnt)))
len(edges_weights)
```

Выбрем пары, у которых вес больше или равен 2

```python
edges_weights = [x for x in edges_weights if x[2] >= 2]
```

## Строим граф

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def draw_graph(edges):
    G = nx.Graph()
    G.add_weighted_edges_from(edges)
    labels = nx.get_edge_attributes(G, "weight")
    pos = nx.spring_layout(G)
    nx.write_gexf( G, "gephi_graph.gexf")
```

```python
draw_graph(edges_weights)
```

В ходе работы был построен граф 50 ключевых (наиболее частотных по корпусу) слов корпуса.
Для этого была сделана предобработка, а именно: токенизация, лемматизация,
нормализация и удаление стоп-слов.
При построении графа учтены следующие условия:

● ключевым словом может быть только существительное

● ключевое слово не может быть стоп-словом

● между двумя словами есть связь, если они встретились в одном документе;
вес ребра между вершинами равен количеству документов, в которых два
рассматриваемых слова встретились вместе

● не включены ребра с весом меньше 2, общее количество вершин равно 50

● для визуализации графа использовался Gephi (в отчете)

● при визуализации настроены размер вершины (по степени) и толщина ребер
(по весу)
